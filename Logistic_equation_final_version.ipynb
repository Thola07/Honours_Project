```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import rcParams

rcParams.update({'font.size': 17, 'font.family': 'STIXGeneral',\
   'mathtext.fontset': 'stix'})
```

```python
#-----------------------------------------------------------------------#
def function(t):
    """
      Function for figure 4 on your book
        P1 = 2.*np.exp(0.7944*t)
        P2 = 64./(1.0+ 31*np.exp(0.7944*t) ) 
      This is a docstring to explain what the function does.
      I will combine the two functions and return them as a tuple
      which is a two dimensional point in analogy. That is 
      return (P1, P2)
  
      The way you get the answer to the exponential at t=1.0 for example
      you call this function(1.0)[0]
    """
    P1 = 2.*np.exp(0.7944*t)
    P2 = 64./(1.0+ 31*np.exp(-0.7944*t) )
    return P1, P2
    # End of the function
```

```python
t = np.arange(0, 17, 1) # The way programming works is that it ignores the 17 and
# ends at 16. Since t is an array it means function(t) will be an array too.
# The 1 in the end is the incrementer. 
# PYTHON or C OR C++ or JAVA use the same counting convention of starting from zero.
# This function works like this np.arange( first_val, last_val, incrementer )
# where incrementer is the same as saying dt.

# Comment the print statement in to confirm
#print "t: ", t
```

```python
P_1, P_2 = function(t) # You can do this since function(t) returns a tuple
                       # OR say 
                       # P_1 = function(t)[0]
                       # P_2 = function(t)[1]
# Comment these next statements in to check the values of P1 and P2 if you want too
#print "P1 : ", P_1
#print "P2 : ", P_2
#-----------------------------------------------------------------------#

#P observed
P_observed = [2., 3., 22., 16., 39., 52., 54., 47.,50., 76.,69.,51.,57.,\
70.,53.,59., 57.]
```

```python
# Lets plot
plt.figure(figsize=(12, 10))
plt.plot(t, P_1, 'k', lw=2, label=r'$2e^{0.7944t}$')
plt.plot(t, P_2, 'm', lw=2, label=r'$\frac{64}{1+e^{-0.7944t}}$')
plt.plot(t, P_observed, 'c.', lw=2, label=r'$P_{\rm observed}$')
plt.ylim(0.0, 70.0)
plt.xlim(0.0, 17.0)
plt.xlabel("t", fontsize=17)
plt.ylabel("P(t)", fontsize=17)
plt.legend(loc="best",prop={'size': 16})
#plt.show()
plt.savefig("fig4numberone.png")
```

```python
# To make the curve smooth for the theorectical plots, decrease the stepsize dt=0.1 
# on the array for t. Then call function(t) again. Notice t is new and also P_1 and P_2
t = np.arange(0, 17, 0.1)
P_1, P_2 = function(t)
plt.figure(figsize=(12, 10))
plt.plot(t, P_1, 'k', lw=2, label=r'$2e^{0.7944t}$')
plt.plot(t, P_2, 'm', lw=2, label=r'$\frac{64}{1+e^{-0.7944t}}$')
plt.ylim(0.0, 70.0)
plt.xlim(0.0, 17.0)
plt.xlabel("t", fontsize=17)
plt.ylabel("P(t)", fontsize=17)
plt.legend(loc="best",prop={'size': 16})
#plt.show()
plt.savefig("fig4numbertwo.png")
```
